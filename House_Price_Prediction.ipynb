```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
price_data = pd.read_csv('/content/House Price India.csv')
```

```python
price_data.shape
```

```python
price_data.head()
```

```python
price_data.describe()
```

```python
price_data.isnull().sum()
```

```python
#removing the columns date an id bcs they are not required..

price_data=price_data.drop(columns=['id','Date'])
```

```python
#making all features as lower case..
price_data.columns = price_data.columns.str.lower()
```

```python
price_data.head()
```

```python
#to visualize the distribution of price
sns.distplot(price_data.price,color='green')
plt.title('Price Distribution')
plt.xlabel('Price')
plt.ylabel('Count')
plt.show()
```

```python
X = price_data.drop(columns=['price'],axis=1)
Y = price_data['price']
```

**Splitting the dataset to Training Data and Testing Data**

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X,Y,test_size=0.2,random_state=42)
```

```python
print(X.shape,X_train.shape,X_test.shape)
```

**Model Evaluation**

```python
from sklearn.metrics import mean_squared_error,mean_absolute_error,r2_score
```

```python
def model_evaluation(true,predicted):
  mse = mean_squared_error(true,predicted)
  mae = mean_absolute_error(true,predicted)
  r2 = r2_score(true,predicted)

  return mse,mae,r2
```

**Model Validation**

```python
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.svm import SVR
from sklearn.neighbors import KNeighborsRegressor
from xgboost import  XGBRegressor
from sklearn.model_selection import cross_val_score
```

```python
models = [LinearRegression(),DecisionTreeRegressor(),RandomForestRegressor(),SVR(),KNeighborsRegressor(),XGBRegressor()]
```

```python
for model in models:
  model.fit(X_train,Y_train)

  #making prediction
  Y_train_pred = model.predict(X_train)
  Y_test_pred = model.predict(X_test)

  #Evaluate train data
  model_train_mse,model_train_mae,model_train_r2 = model_evaluation(Y_train,Y_train_pred)

  #Evaluate test data
  model_test_mse,model_test_mae,model_test_r2 = model_evaluation(Y_test,Y_test_pred)

  print("Model Name:",model)
  print()

  print(f"Model Train MSE:{model_train_mse}")
  print(f"Model Train MAE:{model_train_mae}")
  print(f"Model Train R2:{model_train_r2}")

  print("---------------------------------------------------------------------------")

  print(f"Model Test MSE:{model_test_mse}")
  print(f"Model Test MAE:{model_test_mae}")
  print(f"Model Test R2:{model_test_r2}")

  print("----------------------------------------------------------------------------")

  cv_score = cross_val_score(model,X,Y,cv=5)
  print("Cross Validation:",cv_score)
  print("Average Cross Validation Score:",np.mean(cv_score))

  print("-----------------------------------------------------------------------------")
  print("------------------------------------------------------------------------------")
  print()

```

Here The XGBRegressor has highest score of prediction

**HyperParameter Tuning**

```python
hyperparameters = {
    'XGB' : {
        'n_estimators': [100, 200, 300],
        'max_depth': [3, 5, 7],
        'learning_rate': [0.01, 0.1, 0.2],
        'subsample': [0.7, 0.8, 1.0],
        'colsample_bytree': [0.7, 0.8, 1.0],
        'gamma': [0, 0.1, 0.2]
    }
}
```

**Applying RamdomizedSearchCV**

```python
from sklearn.model_selection import RandomizedSearchCV
```

```python
model_keys = list(hyperparameters.keys())
```

```python
models = [XGBRegressor()]
```

**Model Selection**

```python
def ModelSelection(models_list,hyperparameters,X_train,Y_train):

  result = []
  i=0

  for model in models_list:
    key = model_keys[i]
    params = hyperparameters[key]

    print(model)
    print(params)

    random = RandomizedSearchCV(model,params,cv=5)
    random.fit(X_train,Y_train)

    result.append({
        'model used' : model,
        'highest score' : random.best_score_,
        'best parameters' : random.best_params_
    })
    i+=1

    result_df = pd.DataFrame(result,columns=['model used','highest score','best parameters'])

    return result_df
```

```python
ModelSelection(models,hyperparameters,X_train,Y_train)
```

```python
best_model = XGBRegressor(
    subsample=0.8,
    n_estimators=200,
    max_depth=5,
    learning_rate=0.2,
    gamma=0,
    colsample_bytree=0.8
)
```

```python
best_model.fit(X_train,Y_train)

Y_train_predi = best_model.predict(X_train)
Y_test_predi = best_model.predict(X_test)

model_train_mse,model_train_mae,model_train_r2 = model_evaluation(Y_train,Y_train_predi)

model_test_mse,model_test_mae,model_test_r2 = model_evaluation(Y_test,Y_test_predi)

print(f"Model Train MSE:{model_train_mse}")
print(f"Model Train MAE:{model_train_mae}")
print(f"Model Train R2:{model_train_r2}")

print("---------------------------------------------------------------------------")

print(f"Model Test MSE:{model_test_mse}")
print(f"Model Test MAE:{model_test_mae}")
print(f"Model Test R2:{model_test_r2}")
```

**Visualizing the Actual and Predicted Prices**
